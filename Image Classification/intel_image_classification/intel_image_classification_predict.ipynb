```python
import matplotlib.pyplot as plt
import tensorflow as tf
from PIL import Image
```

```python
model=tf.keras.models.load_model(r'Datasets\models\intelClassification.h5')
```

```python
 
 def predict(image):
    img=Image.open(image)
    plt.imshow(img)
    arr= tf.reshape(tf.divide(tf.constant(img.resize((64,64))),255),(-1,64,64,3))
    labels=('Building','Forest','Glacier','Mountain','Sea','Street')
    preds=model.predict(arr,verbose=0)
    preds=tf.round(preds*100,2).numpy()[0].tolist()
    print(dict(zip(labels,preds)))
    Max=max(preds)
    return f'{labels[preds.index(Max)]} {Max} %'

predict(r'Datasets\Img\intel_image_classification\seg_test\seg_test\mountain\20438.jpg')
```
