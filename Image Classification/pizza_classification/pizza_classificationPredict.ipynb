```python
import matplotlib.pyplot as plt
import tensorflow as tf
from PIL import Image
```

```python
model=tf.keras.models.load_model(r'Datasets\models\pizzaClassification.h5')
```

```python

def predict(image):
    img=Image.open(image)
    plt.imshow(img)
    arr=tf.reshape(tf.divide(tf.constant(img.resize((64,64))),255), (-1,64,64,3))
    preds=model.predict(arr,verbose=0)[0][0]*100
    return f'Pizza {preds} %' if preds>20 else f'Non Pizza {100-preds} %' 

predict(r'Datasets\Img\pizzaClassification\test\not_pizza\3535053.jpg')
```
